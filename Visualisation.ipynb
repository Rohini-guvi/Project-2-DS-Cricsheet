```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=df=pd.read_csv("ipl.csv",index_col=0,low_memory=False)
```

```python
df
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(data = df, y = 'Winner', color = 'y')
plt.title("IPL Winning Teams count")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.histplot(data = df, y = 'City', color = 'g')
plt.title("IPL Matches count held in each locations")
plt.show()
```

```python
df['season'].value_counts().plot(kind='bar', title='IPL Matches played over years', figsize=(8, 5))
```

```python
df1=pd.read_csv("odis.csv",index_col=0,low_memory=False)
```

```python
df=pd.read_csv("t20s.csv",index_col=0,low_memory=False)
```

```python
df.shape
```

```python
df
```

```python
x = df.loc[df.Target_SuperOver!='Super_over']
plt.figure(figsize=(30, 15))
plt.title("T20s matches Target runs against years")
sns.lineplot(data = x, x = df.season, y = x.Target_SuperOver, marker = 'o', color = 'r')
```

```python
df_new=df1[df1['Batter']=="MS Dhoni"]
```

```python
df_new
```

```python
plt.figure(figsize=(15, 5))
plt.title("Performance of MS Dhoni in ODI Matches")
sns.lineplot(data = df_new, x = df_new.Batter_Runs_per_over, y = df_new.Event_Name, marker = 'o', color = 'g')
```

```python
df_new=df[df['Bowler']=="DB Sharma"]
df_new
```

```python
data=pd.DataFrame(df_new.groupby(['Wicket_Kind'])['Bowler'].value_counts())
data
```

```python
plt.figure(figsize=(10,5))
plt.title("DB Sharma's wicket count in T20 matches")
sns.scatterplot(data=data,x='Wicket_Kind',y='count')
```

```python
plt.title("Extra runs added for each Deliveries in DB Sharma's balls")
sns.lineplot(data=df_new,x='Deliveries',y='Extras_per_over')
```
